```python
import matplotlib.pyplot as plt
from arabic_reshaper import ArabicReshaper
from bidi.algorithm import get_display
import pandas as pd
import seaborn as sns


arabic_reshaper = ArabicReshaper()



data['category_reshaped'] = data['category'].apply(lambda x: arabic_reshaper.reshape(x))
data['category_display'] = data['category_reshaped'].apply(lambda x: get_display(x))

plt.rcParams['font.size'] = 12


plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='category_display')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.show()
```

```python
import pandas as pd
from collections import Counter
import arabic_reshaper
from bidi.algorithm import get_display
import matplotlib.pyplot as plt


sentence_counts = Counter(data['comment'])

most_common_sentences = sentence_counts.most_common(5)

sentences, counts = zip(*most_common_sentences)

reshaped_sentences = [get_display(arabic_reshaper.reshape(sentence)) for sentence in sentences]

plt.figure(figsize=(10, 8))
plt.barh(reshaped_sentences, counts, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Sentences')
plt.title('Top 5 Most Repeated Sentences')
plt.gca().invert_yaxis()
plt.show()
```
